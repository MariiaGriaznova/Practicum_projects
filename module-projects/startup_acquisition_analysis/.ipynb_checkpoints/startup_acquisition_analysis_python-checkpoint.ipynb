## Step 2. Data preprocessing, preliminary research


### 2.1. Funding Rounds by Year

This task should be completed without merging or additional preprocessing, based on the `company_and_rounds.csv` dataset.

* Create a pivot table by year, where based on the `raised_amount` column, the following information is provided for each year:

  * The typical amount of funds raised in a single round;
  * The total number of funding rounds in that year.

* Keep only the data for years where there is information on more than 50 funding rounds.

```python
pivot_df = (
    company_and_rounds_df
    .dropna(subset=['funded_at', 'raised_amount']) 
    .assign(year=lambda df: df['funded_at'].dt.year) 
    .groupby('year', as_index=False)
    .agg(
        typical_raised_amount=('raised_amount', 'median'), 
        total_rounds=('funding_round_id', 'count') 
    )
)

# Filter only years with more than 50 rounds
pivot_df = pivot_df[pivot_df['total_rounds'] > 50]

display(pivot_df)
```

Based on the resulting table, we will construct a graph that will reflect the dynamics of the typical amount of funds that startups received within one round of financing.

```python
plt.figure(figsize=(12, 6))
plt.plot(pivot_df['year'], pivot_df['typical_raised_amount'] / 1_000_000, marker='o', linestyle='-')
plt.xlabel('Год')
plt.xticks(pivot_df['year'])
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.ylabel('Typical Funding Size (Millions)')
plt.title('Dynamics of the typical size of funds allocated in one round of financing')

plt.show()
```

- The maximum typical amount of funds raised was in 2005, when it was 5.5 million with a total of 1,633 rounds.

- In 2013, we observe the highest number of rounds (11,072), which is significantly higher than in 2012, where there were 9,970 rounds, and more than in all previous years. This continued growth in the number of rounds may indicate increased interest in startups and more investments, but with smaller amounts raised per round. In 2013, the typical amount raised per round was 1.2 million, which shows a trend towards slowing the decline in funds raised per round (in 2012, this amount was 1 million).

Thus, we can conclude that there has been an increase in the number of smaller startups that are actively being funded. As a result, on one hand, we see growth in the number of rounds and the total investments, and on the other hand, a decline in the amount of funds per investment round.

Before merging, let's check both tables for duplicates.

```python
# Check for duplicates by id
people_duplicates = people_df[people_df.duplicated(subset=['id'], keep=False)]
print(people_duplicates)
```

We see 2 records for Nicole Couto. However, we cannot remove one of the duplicates, as they are incomplete. It appears that this person moved from one company to another, and both companies have records of their education, which is correct for our analysis.

There are 8,001 potentially duplicate records. While for the first 2 records we can confidently say they are duplicates, for records with missing dates in `graduated_at`, we cannot be sure. The `degrees_df` table may assist us here.

Before proceeding with further analysis, we will merge all 3 dataframes into one.

```python
# Rename id columns to avoid confusion
education_df.rename(columns={'id': 'id_education'}, inplace=True)
people_df.rename(columns={'id': 'id_person'}, inplace=True)
degrees_df.rename(columns={'id': 'id_degree'}, inplace=True)

# Remove the 'p:' prefix from 'object_id' and cast to integer type
degrees_df['object_id'] = degrees_df['object_id'].str.replace('p:', '').astype(int)

# Merge tables
people_education = pd.merge(people_df, education_df, how='left', left_on='id_person', right_on='person_id')
people_education_degrees = pd.merge(people_education, degrees_df, how='left', left_on='id_person', right_on='object_id')

# Let's check the result
people_education_degrees.head()
```

```python
people_education_degrees.info()
```

```python
# Check for complete duplicates
people_education_degrees[people_education_degrees.duplicated(keep=False)]
```

Next, to find duplicates, we will use a strict approach that takes into account all the important parameters related to the employee’s education.

```python
duplicates = people_education_degrees[people_education_degrees.duplicated(subset=['id_person', 'company_id', 'institution', 'graduated_at', 'id_degree', 'object_id', 'degree_type', 'subject'], keep=False)].sort_values(by=['id_person', 'id_degree'])
display(duplicates)
```

```python
# Remove duplicates from the original dataframe using indices from the duplicates variable
people_education_degrees = people_education_degrees.drop(index=duplicates.index)

display(people_education_degrees)
```

After removing duplicates, the number of rows in the `people_education_degrees` dataset was reduced from 370,836 to 349,746 records.

### 2.3. To Merge or Not to Merge — That Is the Question

Some column names appear more frequently across datasets. For example, during the preliminary check of the datasets, it was found that the `company_id` column is suitable for merging data, as most of its values appear in multiple datasets more than once. This column can be used for merging.

* Determine if the `network_username` column, which appears in several datasets, is suitable for merging. We need to understand whether information in columns with this name is duplicated across datasets, and if so, how frequently.

* Assess whether the columns with the name `network_username` can be used for merging data.

We know that the `network_username` column appears in the following datasets:
- `company_and_rounds` — company network handle.
- `people` — social media handle (likely the employee's handle, not the company's).
- `fund` — fund network handle.

At first glance, it seems that this column may not be suitable for merging tables, as it should contain different information. However, let's verify this.

First, we will count the number of unique `network_username` values in each dataset and display the first 10 unique values sorted alphabetically to visually compare them across the tables.

```python
print("Unique network_username in company_and_rounds:", company_and_rounds_df["network_username"].nunique())
print("Unique network_username in people:", people_df["network_username"].nunique())
print("Unique network_username in fund:", fund_df["network_username"].nunique())

display("Examples of network_username in company_and_rounds:")
display(sorted(company_and_rounds_df["network_username"].dropna().unique())[:10])  # Первые 10 отсортированных уникальных значений

display("\nExamples of network_username in people:")
display(sorted(people_df["network_username"].dropna().unique())[:10])

display("\nExamples of network_username in fund:")
display(sorted(fund_df["network_username"].dropna().unique())[:10])
```

The number of unique `network_username` values varies across the datasets. In `company_and_rounds`, the maximum number of unique values is 79,571, in `people` it is 38,421, and the smallest number is in `fund` with 2,098.

It is evident that overlaps in values are possible. For example, in both `company_and_rounds` and `people`, the `network_username` field contains the handle '077football'.

Now, let's check if there are overlaps between the `network_username` values from different tables.

```python
# Create sets of unique values from each dataset
company_usernames = set(company_and_rounds_df["network_username"].dropna().unique())
people_usernames = set(people_df["network_username"].dropna().unique())
fund_usernames = set(fund_df["network_username"].dropna().unique())

# Checking intersections
people_in_company = people_usernames.intersection(company_usernames)
fund_in_company = fund_usernames.intersection(company_usernames)
all_three_intersection = people_usernames.intersection(company_usernames, fund_usernames)

# Calculate percentages
people_in_company_percent = (len(people_in_company) / len(company_usernames)) * 100
fund_in_company_percent = (len(fund_in_company) / len(company_usernames)) * 100
all_three_in_company_percent = (len(all_three_intersection) / len(company_usernames)) * 100

# Output the results
print(f"network_username from people that are in company_and_rounds: {len(people_in_company)} ({people_in_company_percent:.2f}%)")
print(f"network_username from fund, which are in company_and_rounds: {len(fund_in_company)} ({fund_in_company_percent:.2f}%)")
print(f"network_username, common to all three tables: {len(all_three_intersection)} ({all_three_in_company_percent:.2f}%)")
```

```python
venn3([company_usernames, people_usernames, fund_usernames], set_labels=('Company Users', 'People Users', 'Fund Users'))

plt.title("Usernames intersection")
plt.show()
```

Based on the results:

- In 2.76% of cases, `network_username` from `people` matches with `network_username` from `company_and_rounds`. This corresponds to 2,199 intersections.
- In 0.09% of cases, `network_username` from `fund` matches with `network_username` from `company_and_rounds`. This corresponds to 70 intersections.

In total, **7 `network_username`** appear in all three tables, which accounts for **0.01%** of the total number of unique values in each dataset.

Thus, **the intersection between the tables based on `network_username` is very minimal**, indicating that if data merging is required, **it is better to consider using other columns** that provide a more accurate and complete connection between the datasets, such as `company_id`.

Calculate the boundaries of the expected outliers using the interquartile range (IQR) method

```python
Q1 = filtered_companies['funding_total_mln'].quantile(0.25)
Q3 = filtered_companies['funding_total_mln'].quantile(0.75)
IQR = Q3 - Q1

# Boundaries for emissions
# Calculate emission boundaries with rounding to 2 digits
lower_bound = round(Q1 - 1.5 * IQR, 2)
upper_bound = round(Q3 + 1.5 * IQR, 2)

# If the lower bound is negative, replace it with 0
if lower_bound < 0:
    lower_bound = 0

# Output the boundary values
display(f"Emission limits: higher {upper_bound}")

# Filtering emissions
outliers = filtered_companies[(filtered_companies['funding_total_mln'] > upper_bound)]

display(f"Number of outliers: {outliers.shape[0]}")

# Check percentile for upper bound
upper_percentile = (filtered_companies['funding_total_mln'] <= upper_bound).mean() * 100

# Output the upper bound percentile
display(f"Upper bound percentile {upper_bound}: {upper_percentile:.2f}%")
```

All funding amounts above 14.13 million are considered potential outliers - this is 14.38% of all companies, but it cannot be ruled out that these large amounts were actually invested in companies. We will plot graphs without extreme funding values for greater clarity, but we will not remove high funding amounts from further analysis.
